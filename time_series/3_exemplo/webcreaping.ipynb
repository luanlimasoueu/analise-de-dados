```python
import urllib3

url = "https://www.agrolink.com.br/cotacoes/historico/sc/frango-1kg"


conexao = urllib3.PoolManager()

retorno = conexao.request('GET',url)

from bs4 import BeautifulSoup

pagina = BeautifulSoup(retorno.data, 'html.parser')
```

```python
print(pagina.prettify())
```

```python
pagina.table
```

```python
tabela = pagina.find_all('table', class_= "table table-main orange")
```

```python
tabela
```

```python
dado = []
for celulas in tabela:
    cabecalho = celulas.find_all(['td','th'], class_= "text-center")
    for dados in cabecalho:
        print(dados)
        dado.append(dados.find(text=True))
```

```python
n = 241
splited = []
len_l = len(dado)
for i in range(n):
    start = int(i*len_l/n)
    end = int((i+1)*len_l/n)
    splited.append(dado[start:end])
print(splited)
```

```python
cabecalho = splited[0]
```

```python
cabecalho
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(splited[1:])
```

```python
df.columns = splited[0]
```

```python
df
```

```python
df['Mês/Ano'] = pd.to_datetime(df['Mês/Ano'])
```

```python
df = df.set_index('Mês/Ano')
```

```python
df['Estadual']= df['Estadual'].str.replace(',', '.').astype(float)
df['Nacional'] = df['Nacional'].str.replace(',', '.').astype(float)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

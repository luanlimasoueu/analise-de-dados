```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.abs(np.random.randn(50)*100)}
data['b'] = 1.5*np.sqrt(data['a']) + np.random.randn(50)
```

```python
fig, ax = plt.subplots(figsize=(8, 5))
ax.scatter('a', 'b',   # variveis a serem plotadas
            c='c',     # definio das cores dos marcadores
            s='d',     # defnio do tamanho dos marcadores
            data=data  # fonte dos dados
          )
ax.set_xlabel('Valores de $a$')                    # Rtulo do eixo X
ax.set_ylabel('Valores de $\sqrt{a}$ com rudo');  # Rtulo do eixo Y
```

```python

# Implementation of matplotlib function  
import matplotlib.pyplot as plt  
import numpy as np  
from matplotlib.ticker import EngFormatter  
    
val = np.random.RandomState(19680801)  
xs = np.logspace(1, 9, 100)  
ys = (0.8 + 4 * val.uniform(size = 100)) * np.log10(xs)**2
     
plt.xscale('log')  
plt.plot(xs, ys)  
plt.xlabel('Frequency')  
    
plt.title('matplotlib.pyplot.xscale() function Example\n', fontweight ="bold")  
    
plt.show()  
```

```python
# Implementation of matplotlib function  
import numpy as np  
import matplotlib.pyplot as plt  
    
fig, ax4 = plt.subplots()  
    
x = 10.0**np.linspace(0.0, 2.0, 15)  
y = x**2.0
plt.xscale("log")  
plt.yscale("log")  
    
plt.errorbar(x, y, xerr = 0.1 * x,  
             yerr = 2.0 + 1.75 * y,   
             color ="green")  
    
plt.ylim(bottom = 0.1)  
    
plt.title('matplotlib.pyplot.xscale() function Example\n', fontweight ="bold")  
    
plt.show()  
```

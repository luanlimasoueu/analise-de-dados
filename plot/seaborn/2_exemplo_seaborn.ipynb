```python
import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns

%matplotlib inline

sns.set(style="darkgrid")
```

```python
gorjetas = sns.load_dataset("tips")

gorjetas.head()
```

```python
plt.figure(figsize=(8,5))

plt.title('Valor da conta X Valor da gorjeta')

sns.regplot(x="total_bill", y="tip", data=gorjetas)

plt.show()
```

```python
sns.lmplot(x="total_bill", y="tip", col = 'day', data=gorjetas)
```

```python
sns.jointplot(x="total_bill", y="tip", kind = 'reg', data=gorjetas)
```

```python
plt.figure(figsize=(8,5))

plt.title('Valor da conta e Valor da gorjeta para M/F')

sns.scatterplot(x="total_bill", y="tip", hue = 'sex', data=gorjetas)

plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title('Valor da gorjeta por dia')
sns.boxplot(x="tip", y="day", data=gorjetas)
plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.title('Idade de passageiros em cada classe por sexo')

sns.violinplot(x="day", y="tip", hue = 'sex', split = True, data=gorjetas)

plt.show()
```

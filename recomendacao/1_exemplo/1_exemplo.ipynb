```python
import pandas as pd
dicionario = {
    'SKU': ['C1', 'C1', 'C3', 'C2'],
    'DATA': ['2018-01', '2018-07', '2019-07', '2019-07'],
    'RECEITA': [20, 30, 10, 5],
}
df = pd.DataFrame(dicionario)


lista_mes_ano = []
for ano in range( 2018, 2024):
    for mes in range (1, 13):
        ano = str(ano)
        if mes <10:
            mes = '0'+str(mes)
        else:
            mes = str(mes)
        periodo = ano+'-' +mes
        lista_mes_ano.append(periodo)



lista_adicao = []
for i in list(df['SKU'].unique()):
    df_temp = df[df['SKU'] == i]
    for j in lista_mes_ano: 
        if j not in df_temp['DATA'].unique():
            lista_adicao.append([i, j, 0])
                
df_join = pd.DataFrame( lista_adicao, columns=['SKU',  'DATA' , 'RECEITA'])
df =  pd.concat([df, df_join], ignore_index=True)
```
